```python
# The selected database provides estimates of number of relevant children 5 to 17 years old in poverty across school districts in the US
```

```python
df = pd.read_csv("school_povertyestimates.csv") 
df
```

```python
df['Estimated number of relevant children 5 to 17 years old in poverty '] = pd.to_numeric(df['Estimated number of relevant children 5 to 17 years old in poverty '])
```

```python
mean_value = df['Estimated number of relevant children 5 to 17 years old in poverty '].mean()
median_value = df['Estimated number of relevant children 5 to 17 years old in poverty '].median()
mode_value = df['Estimated number of relevant children 5 to 17 years old in poverty '].mode()[0]

print(f"Mean of the number of relevant children 5 to 17 years old in poverty across all school districts is: {mean_value}")
print(f"Median of the number of relevant children 5 to 17 years old in poverty across all school districts is: {median_value}")
print(f"Mode of the number of relevant children 5 to 17 years old in poverty across all school districts is: {mode_value}")
```

```python
import csv
from statistics import mean, median, mode
```

```python
with open('school_povertyestimates.csv', 'r') as file:
    reader = csv.DictReader(file)
    povertydata_2020 = [(row['Name'], int(row['Estimated number of relevant children 5 to 17 years old in poverty '])) for row in reader]

maximum_barheight = 2000
highest_value = max(value for school,value in povertydata_2020)

for school,value in povertydata_2020:
    barheight = int((value/highest_value) * maximum_barheight) #Maximum height of each bar determined in terms of '#'
    print(f"{school}: " + "#" * barheight) 
    
```
